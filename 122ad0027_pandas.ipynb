```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('car_crashes');
df.head(5)
df1=df.head(30)
df1.head(3)
```

```python
# barh plot of total vs abbrev
df1.head(10).plot(kind='barh',x='abbrev',y='total',color=['aquamarine','seagreen','green'])
plt.show()
```

```python
# add multiple box plots of total vs abbrev
df1[['total','speeding','alcohol']].plot(kind='box',figsize=(6,6),color='red',notch=True, patch_artist = True, vert = 1, boxprops=dict(facecolor='seagreen'))

```

```python
df1['total'].plot(kind='kde',figsize=(6,6),color='red',linewidth=1,linestyle='dotted',marker='o')
plt.title('KDE plot of total')
plt.xlabel('total')
plt.show()
```

```python
#  plots pandas area plot
df1.plot.area(figsize=(10,10),stacked=False)
plt.show()
```

```python
# area plot only on not_distracted and ins_premium
df1[['total','speeding','alcohol']].plot.area(figsize=(10,6),stacked=False)
plt.show()
```

```python
# density plot
df1[['total','speeding','alcohol']].plot.density(figsize=(10,6))
plt.title('Density plot of total, speeding and alcohol')
plt.show()
```

```python
# hexbin plot
df1.plot.hexbin(x='total',y='speeding',gridsize=10,figsize=(10,6))
plt.show()
```
