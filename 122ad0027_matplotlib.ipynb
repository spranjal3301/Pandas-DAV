```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('car_crashes');
df.head(5)
df1=df.head(30)
df1.head(3)
```

```python
x=df['speeding']
y=df['alcohol']
plt.scatter(x,y)
plt.xlabel('speeding')
plt.ylabel('alcohol')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
plt.subplots_adjust( wspace=1, hspace=1)
ax1.hist(df['speeding'])
ax1.set_title('speeding')
ax2.plot(df['no_previous'],df['total'],'ro')
ax2.set_title('no_previous')
ax3.scatter(df['speeding'],df['total'])
ax3.set_title('speeding vs total')
ax4.plot(df['speeding'],color='green',linestyle='--',marker='o',markerfacecolor='red')
ax4.set_title('speeding')
plt.show()
```

```python
plt.plot(df1['not_distracted'],linestyle=':',color='red');
plt.title('not_distracted')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(df1['alcohol'], color="black", linestyle="dashed", label="Speeding");
ax.plot(df1['speeding'], color="red", linestyle="dashed",drawstyle="steps-post", label="alcohol");
# ax.set_title('alcohol vs speeding')
ax.annotate("speeding",xy=(11,8),xytext=(21,8),arrowprops=dict(facecolor="red", headwidth=4, width=1,
                                headlength=4),horizontalalignment="right", verticalalignment="top",weight="bold",color="red");
ax.annotate("alcohol",xy=(25,9),xytext=(20,9),arrowprops=dict(facecolor="red", headwidth=4, width=1,
                                headlength=4),horizontalalignment="right", verticalalignment="top",weight="bold");

ax.legend();
```

```python
# bar plot of total vs abbrev
plt.bar(df1.head(15)['total'],df1.head(15)['abbrev'],color='green',width=0.5)
plt.xlabel('total')
plt.ylabel('abbrev')
plt.show()
```

```python
# pie plot of percentage of speeding and add label to it

lab=df1.head(5)['abbrev']
c=['green','olivedrab','rosybrown','maroon']
plt.pie(df1.head(5)['speeding'],labels=lab,autopct='%1.1f%%',explode=[0,0,0,0,0.1],shadow=True,colors=c)

# plt.pie(df1.head(5)['alcohol'],labels=df1.head(5)['alcohol'],autopct='%1.1f%%')

plt.show()

```

```python
# box plot of speeding
plt.boxplot(df1['speeding'], notch=True, patch_artist = True, vert = 1, boxprops=dict(facecolor='yellow'))
plt.show()
```
