```python

# ffill :propagating the last observed non-null value forward along a specified axis.
# bfill: propagating the next observed non-null value backward along a specified axis.
df = pd.DataFrame({'A': [1, np.nan, 3, np.nan, 5],
                   'B': [np.nan, 2, np.nan, 4, np.nan]})

print("Original DataFrame:")
print(df)

# Forward fill missing values along the rows (axis=0)
filled_df = df.ffill(axis=0)

print("\nDataFrame after forward fill (axis=0):")
print(filled_df)

# Forward fill missing values along the columns (axis=1)
filled_df_axis1 = df.ffill(axis=1)

print("\nDataFrame after forward fill (axis=1):")
print(filled_df_axis1)

# limit: limit the number of consecutive missing values that are filled in ffill or bfill.
filled_df_axis1_limit = df.ffill(axis=1, limit=2)

# Forward fill missing values with a tolerance of 1(max distance  between non-null and null will be fill).
# filled_df = df.ffill(tolerance=1)


```

```python
# fill_value : Value to use for missing values. Defaults to NaN.
data.reindex(columns=['id','diet','pulse','time','kind','random'],fill_value=0)

# NaN 
data.reindex(columns=['id','diet','pulse','time','kind','random'],fill_value=None)

#level: use in matching level in MultiIndex.
data.reindex(columns=['id','diet','pulse','time','kind','random'],level=1)

#copy: If True, returns a copy of the underlying data. If False, a view is returned. Defaults to False.
data.reindex(columns=['id','diet','pulse','time','kind','random'], level=1, copy=True)





```

```python

#* loc: primarily label-based indexing, which means it is used to access data using the index labels or row/column names.
#* iloc: primarily integer-based indexing, which means it is used to access data using the integer indices.
```

```python
#df[column_name]
data['diet']

#! loc
#*Syntax: df.loc[row_label, column_label] or df.loc[row_label]
print(data.loc[2])
print(data.loc[1:4,'diet']) # [:,'diet'] whole column

#! iloc
#*Syntax: df.iloc[row_index, column_index] or df.iloc[row_index]
print(data.iloc[1:4,2]) # [:,0] whole column
print(data.iloc[1:4,:]) # [:,0] whole row

# df.at[row, col]	Select a single scalar value by row and column label
print(data.at[1,'diet']) #[row_label, column_label]

# df.iat[row, col]	Select a single scalar value by row and column position (integers) index
print(data.iat[1,2]) #[row_index, column_index]

```

```python

# Flexible arithmetic methods
# Method	Description
# add, radd	Methods for addition (+)
# sub, rsub	Methods for subtraction (-)
# div, rdiv	Methods for division (/)
# floordiv, rfloordiv	Methods for floor division (//)
# mul, rmul	Methods for multiplication (*)
# pow, rpow	Methods for exponentiation (**)

print(data['Unnamed: 0'].add(data['id']))
# radd : reverse addition  (second DataFrame are added to the values from the first DataFrame.)
# result = data['Unnamed: 0'].radd(data['id'])

print(data['Unnamed: 0'].sub(data['id']))

print(data['Unnamed: 0'].div(data['id']))

# round of division
print(data['Unnamed: 0'].floordiv(data['id']))

print(data['Unnamed: 0'].mul(data['id']))

print(data['Unnamed: 0'].pow(data['id']))
```

```python

#!  rank() based on the position of an element within the sorted dataset
# tie-breaking methods are used to determine the rank of values that are equal. 


# "average"	Default: assign the average rank to each entry in the equal group
data['Avg Rank']=data['pulse'].rank(method='average')

# "min"	Use the minimum rank for the whole group
data['Min Rank']=data['pulse'].rank(method='min')

# "max"	Use the maximum rank for the whole group
data['Max Rank']=data['pulse'].rank(method='max')

# "first"	Assign ranks in the order the values appear in the data
data['First Rank'] = data['pulse'].rank(method='first')

# "dense"	Like method="min",  rank data similar to "min", but the ranks are always consecutive integers without any gaps.
data['Dense Rank'] = data['pulse'].rank(method='dense')

print(data.head(3))
```
